# SORU 1

Bir string’in boş (empty) olup olmadığını kontrol eden bir fonksiyon yazınız. Fonkiyonun adı `is_empty` olsun. 

````python
def is_empty(string_):
    return len(string_)==0 
````

````python
print(is_empty(""))
````

````python
print(is_empty(""))
````

````python
print(is_empty("Hello"))
````

# SORU 3

Bir sözcüğün (word) palindrom (palindrome) olup olmadığını tespit eden bir fonksiyon yazınız. İngilizce dizi için geçerlidir.

Bir palindrom, tersten okunduğunda da aynı olan kelimedir; örneğin "madam" veya hello" gibi.

````python
def is_palindrome(string_):
    return string_ ==string_[::-1]
````

````python
print(is_palindrome("deneme"))
````

````python
print(is_palindrome("efe"))
````

````python
print(is_palindrome("madam"))
````

````python
print(is_palindrome("hello"))
````

# SORU 8

Bir öğrencinin notuna göre harf notunu döndüren `harf_notu` isimli bir fonksiyon yazınız.

- 90-100 → A,
- 80-89 → B, 
- 70-79 → C, 
- 60-69 → D,
- 0-59 → F

Uygun olmayan bir sayı girildiğinde `"Geçersiz puan"` ifadesi ekrana yazdırılsın.

````python
def harf_notu(puan):
    if 90<=puan<=100:
        return "A"
    elif 80 <=puan<=89:
        return "B"
    elif 70<=puan<=79:
        return "C"
    elif 60<=puan<=69:
        return "D"
    elif 0<=puan<=59:
        return "F"
    else:
        return "Geçersiz puan"
````

````python
print(harf_notu(85))
````

````python
print(harf_notu(102))
````
